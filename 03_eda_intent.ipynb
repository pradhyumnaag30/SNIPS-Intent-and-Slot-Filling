```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

pd.set_option("display.max_colwidth", 200)
```

```python
train_df = pd.read_csv("dataset/snips_intent_train.csv")
test_df = pd.read_csv("dataset/snips_intent_test.csv")
```

```python
print("Train shape:", train_df.shape)
print("Test shape:", test_df.shape)
print("Unique intents Train:", train_df['intent'].nunique())
print("Unique intents Test:", test_df['intent'].nunique())

print("\nNull values:")
print(train_df.isnull().sum())
print(test_df.isnull().sum())
```

```python
plt.figure(figsize=(10,5))
train_df['intent'].value_counts().plot(kind='bar')
plt.title("Intent Distribution (Train)")
plt.xlabel("Intent")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
test_df['intent'].value_counts().plot(kind='bar')
plt.title("Intent Distribution (Test)")
plt.xlabel("Intent")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
train_df['text_length'] = train_df['text'].apply(lambda x: len(x.split()))

plt.figure(figsize=(8,5))
sns.histplot(train_df['text_length'], bins=20, kde=True)
plt.title("Text Length Distribution in words (Train)")
plt.xlabel("Words per text")
plt.show()

train_df['text_length'].describe()
```

```python
test_df['text_length'] = test_df['text'].apply(lambda x: len(x.split()))

plt.figure(figsize=(8,5))
sns.histplot(test_df['text_length'], bins=20, kde=True)
plt.title("Text Length Distribution in words (Test)")
plt.xlabel("Words per text")
plt.show()

test_df['text_length'].describe()
```

```python
for intent in train_df['intent'].unique():
    print("\n---", intent, "---")
    print(train_df[train_df['intent']==intent].sample(2)['text'].values)
```

```python
for intent in test_df['intent'].unique():
    print("\n---", intent, "---")
    print(test_df[test_df['intent']==intent].sample(2)['text'].values)
```
